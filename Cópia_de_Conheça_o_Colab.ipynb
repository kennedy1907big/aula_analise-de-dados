```
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Paçoquinha", "Diana", "Laura"],
    "nota": [8.5, 6.0, 7.2, 5.5, 9.0],
    "faltas": [5, 4, 8, 6, 6]

}

df = pd.DataFrame( dados)
print(df)

media_nota = df["nota"].mean()
print("Média das notas:", media_nota)

mais_de_5_faltas = df[df["faltas"] > 5]
print(mais_de_5_faltas)

ordenado = df.sort_values(by="nota", ascending=False)
print(ordenado)

total_faltas = df["faltas"].sum()
print("Total de faltas", total_faltas)

df["sitaucao"] = df["nota"].apply(
    lambda x: "Aprovado" if x >= 7 else "Reprovado"
)
print(df)
```
