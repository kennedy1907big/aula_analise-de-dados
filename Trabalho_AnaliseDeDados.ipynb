```
import pandas as pd


df = pd.read_excel('dataset_salarios_empresa.xlsx')
df.head() #Exibe as primeiras linhas
```

```
df["Salario_R$"].mean()
```

```
df.groupby("Departamento")["Salario_R$"].mean()
```

```
print("Maior Salario:", df["Salario_R$"].max())
print("Menor Salario:", df["Salario_R$"].min())
```

```
custo_departamento = df.groupby("Departamento")["Salario_R$"].sum()
print(custo_departamento)
print("Departamento com maior custo:", custo_departamento.idxmax())
```

```
df["Departamento"].value_counts()
```

```
custo_hora = df["Salario_R$"].sum() / 21
custo_hora = custo_hora / 8
print(custo_hora)
```

```
top10 = df.sort_values(by="Salario_R$", ascending=False).head(10)
print(top10)
```

```
df.groupby("Cargo")["Salario_R$"].mean()
```

```
df.groupby("Departamento")["Idade"].mean()
```

```
import matplotlib.pyplot as plt
```

```
plt.hist(df["Salario_R$"], bins=15)
plt.title("Distribuição Salarial")
plt.xlabel("Salario")
plt.ylabel("Quantidade de Funcionários")
plt.show()
```
